### 1. Importing libraries

```python
import errno
import glob
import itertools
import nltk
import random
import requests
import shutil

from collections          import Counter
from itertools            import *
from IPython.display      import HTML, display
from nltk.corpus          import stopwords
from nltk.probability     import FreqDist
from nltk.stem.porter     import PorterStemmer
from nltk.stem.wordnet    import WordNetLemmatizer
from nltk.tokenize        import word_tokenize
from random               import randint
from urllib               import request
from wordcloud            import WordCloud


import matplotlib.pyplot  as plt
import numpy              as np
import pandas             as pd
```

### 2. Reading 20 *.txt files splitting them into strings and excluding stopwords

```python
# Read in one file and split into strings
path = 'books/en_books/*.txt'
files = glob.glob(path)
for sentence in files:
    with open(sentence, encoding="utf-8") as fh:
        booksen = fh.readlines()
        booksenstr = [word for line in booksen for word in line.split()]
```

```python
# Convert charList to a string.
finalString = ''.join(booksenstr)
```

```python
# Remove stopwords in wordcloud
example_sent = finalString
  
stop_words = set(stopwords.words('english')) 
  
word_tokens = word_tokenize(example_sent) 
  
parola = [w for w in word_tokens if not w in stop_words] 
  
parola = []
for w in word_tokens: 
    if w not in stop_words: 
        parola.append(w)
```

### 3. Generate a Wordcloud

##### What is a wordcloud?
A word cloud is a collection, or cluster, of words depicted in different sizes. The bigger and bolder the word appears, the more often it’s mentioned within a given text and the more important it is.
##### What is done here?
Here a word cloud is created out of the "All_in_one_en.txt" to show if a word is mentioned more often in a text than other ones.

```python
UNIQUE_STRING = (" ").join(parola)
WORDCLOUD = WordCloud(width=1000, height=500).generate(UNIQUE_STRING)
plt.figure(figsize=(30, 16))
plt.imshow(WORDCLOUD)
plt.axis("off")
plt.show()
```

### 4. Histogram

##### What is a histogram?
Histogram is a chart representing a frequency distribution; heights of the bars represent observed frequencies.
##### What is done here?
This histogram here shows the five most frequently appearing words in "All_in_one_en.txt" and even shows the number of appearances.

```python
vocab_labels, vocab_values = zip(*Counter(booksenstr).items())

# Sort and fix values and labels
sorted_values = sorted(vocab_values)[::-1]
sorted_labels = [x for (y,x) in sorted(zip(vocab_values,vocab_labels))][::-1]  
indexes = np.arange(len(sorted_labels[:5]))                                    
width = 1                                                                                                                                                

# Design figure
plt.figure(figsize=(15, 5))
plt.bar(indexes, sorted_values[:5])
plt.xticks(indexes + width * 0.1, sorted_labels[:5])                                                                                    
plt.show()
```

### 7. Word Stemming

##### What is a word stemming?
Lemmatization is the process of grouping together the different inflected forms of a word so they can be analysed as a single item.
##### What is done here?
Here one word out of 200 from "All_in_one_en.txt" is randomly chosen and a stemma is generated from it.

```python
lem = WordNetLemmatizer()

stem = PorterStemmer()

word = booksenstr[randint(0, 200)]
print("Lemmatized Word:",lem.lemmatize(word,"v"))
print("Stemmed Word:",stem.stem(word))
```

### 8. Part of Speech Tagging

##### What is a part of speech tagging (POST)?
Part-of-speech tagging is the process of marking up a word in a text (corpus) as corresponding to a particular part of speech, based on both its definition and its context — i.e., its relationship with adjacent and related words in a phrase, sentence, or paragraph. A simplified form of this is commonly taught to school-age children, in the identification of words as nouns, verbs, adjectives, adverbs, etc.
##### What is done here?
A random choosen group of words is chosen out of the first 1000 ones in the "All_in_one_en.txt". Afterwards all the contained words are going to be tagged.               		

```python
for _ in range(1000):
    value = randint(0, 1000)

sent = booksen[value]

tokens=nltk.word_tokenize(sent)
print(tokens)

nltk.pos_tag(tokens)
```
